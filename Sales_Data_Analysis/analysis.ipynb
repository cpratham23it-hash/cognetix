```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")
df.head()
df.info()
df.describe()
df = df.dropna()
df = df.drop_duplicates()
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"])
total_sales = df["SALES"].sum()
avg_sales = df["SALES"].mean()
max_sale = df["SALES"].max()
print("Total Sales:", total_sales)
print("Average Sales:", avg_sales)
print("Max Sale:", max_sale)
df.groupby("ORDERNUMBER")["SALES"].sum()
df.groupby("CITY")["SALES"].sum()
```

```python
df.groupby("CITY")["SALES"].sum().plot(kind='bar')
```

```python
df.groupby("STATE")["SALES"].sum().plot(kind='pie')
```

```python
df.resample("M", on="ORDERDATE")["SALES"].sum().plot()
```

```python
df.groupby("PRODUCTCODE")["SALES"].sum().nlargest(10).plot(kind='bar')
```
